<summary>Git, Github 튜토리얼</summary>

git과 GitHub에 대해 알아야 할 내용을 정리해 드리겠습니다.

**Git**:
*   **분산 버전 관리 시스템**: 코드 변경 사항을 추적하고 관리하는 데 사용되며, 개발자 각자의 컴퓨터에 전체 코드 베이스와 변경 이력을 저장합니다. 이를 통해 여러 개발자가 동시에 작업하고 변경 사항을 추적하며 이전 버전으로 쉽게 되돌릴 수 있습니다.
*   **`git version`**: 설치된 Git의 버전을 확인합니다.
*   **`git config --global user.name '이름'`**: 전역 환경에 깃을 변경한 사람의 이름을 설정합니다. 재변경하려면 '이름' 위치의 값만 바꾸면 됩니다.
*   **`git config --global user.email '이메일 주소'`**: 전역 환경에 깃을 변경한 사람의 이름을 설정합니다. 재변경하려면 '이메일 주소' 위치의 값만 바꾸면 됩니다.
*   **로컬 저장소**:  `git init` 명령어를 사용하여 프로젝트 폴더 내에 생성되며, 변경 사항은 로컬에 저장됩니다.
    * 최초에 master로 경로가 나올텐데, 최근에는 main이라는 말을 더 많이 씀. **`git branch -m main`**으로 변경하면 됨.
    * 이후로는 계속 main으로 하고 싶다면 다음 명령어를 실행하면 됨. **`git config --global init.defaultBranch main`** 
*   **커밋**: 코드 변경 사항을 저장하는 스냅샷으로, 특정 시점의 프로젝트 상태를 기록합니다. 커밋 메시지를 작성하여 변경 사항을 설명하는 것이 좋습니다.
    * **`git status`**: 깃 상태를 보여줌
*   **브랜치**: 프로젝트의 병렬 버전으로, 여러 기능을 동시에 개발할 수 있게 해줍니다. 기본 브랜치 이름은 과거에는 `master`였지만 현재는 `main`을 주로 사용합니다.
*   **`git add`**: 변경된 파일을 Git이 추적할 수 있도록 스테이징 영역에 추가합니다.
        git add readme.md
*   **`git commit`**: 스테이징 영역의 변경 사항을 저장합니다. 변경사항은 주기적으로 저장해줘야 함

    <summary>commit은 얼마나 자주해야 하는지?</summary>
    `git commit`을 얼마나 자주 해야 하는지, 그리고 어떤 관점에서 해야 하는지에 대한 질문이시군요. `git commit`은 프로젝트의 변경 사항을 기록하는 중요한 작업이며, **너무 자주 하면 번거롭고, 너무 드물게 하면 변경 사항 추적이 어려워질 수 있습니다.** 따라서 적절한 커밋 주기를 설정하는 것이 중요합니다.

    **커밋 빈도 및 관점**:

    *   **정기적인 커밋의 중요성**: 소스에 따르면, **정기적인 커밋은 프로젝트의 진행 상황을 추적하고, 문제가 발생했을 때 이전 버전으로 되돌아가기 쉽게 해줍니다**. 커밋은 프로젝트의 특정 시점의 "스냅샷"과 같아서, 필요할 때 언제든지 과거의 코드로 돌아갈 수 있게 해줍니다.
    *   **기능 단위 커밋**: 소스에 따르면, **각 커밋은 하나의 논리적인 변경 단위**를 나타내야 합니다. 즉, **하나의 기능 추가, 버그 수정, 코드 개선 등**과 같은 의미 있는 변경 사항을 묶어서 커밋해야 합니다. 이렇게 하면 변경 사항을 추적하고 관리하기가 더 쉬워집니다.
    *   **커밋 메시지**: 소스에 따르면, 커밋 메시지를 **명령형(imperative mood)**으로 작성해야 합니다. 이는 마치 명령을 내리는 것처럼 작성하는 것을 의미하며, **"이 커밋을 적용하면 무엇이 변경되는가?"**에 대한 답을 명확하게 제시해야 합니다. 예를 들어, "모바일 반응성 개선", "AB 테스트 추가"와 같이 작성하는 것이 좋습니다.
    *   **커밋 내용**: 커밋에는 **의미 있는 변경 사항**이 포함되어야 합니다. 소스에서와 같이, "readme.md 파일 추가"와 같이 어떤 파일을 추가했는지, 어떤 내용을 수정했는지에 대한 정보를 담아야 합니다.
    *   **너무 잦은 커밋**: 소스에는 명시적으로 언급되어 있진 않지만, 너무 작은 변경 사항마다 커밋을 하면 커밋 로그가 복잡해지고, 변경 사항을 추적하기 어려워질 수 있습니다. 작은 변경들을 **논리적인 단위로 묶어서 커밋**하는 것이 좋습니다.
    *   **너무 드문 커밋**: 소스에 따르면, 커밋을 너무 드물게 하면 **문제가 발생했을 때 이전 버전으로 되돌아가기 어렵고**, 여러 개발자가 동시에 작업할 때 **충돌이 발생하기 쉽습니다.** 자주 커밋하여 변경 사항을 저장하는 습관을 기르는 것이 좋습니다.

    **커밋 주기 요약**:
    *   **기능 구현 완료 시점**: 하나의 기능을 구현하거나 버그를 수정했을 때 커밋합니다.
    *   **코드의 의미 있는 변경 시점**: 코드의 변경이 논리적인 단위를 이루었을 때 커밋합니다.
    *   **작업 흐름 중간 저장**: 작업을 진행하다가 잠시 멈춰야 하거나, 다른 작업을 해야 할 때 현재까지의 작업 내용을 커밋합니다. 이 때, 아직 완료되지 않은 작업은 '작업 중'이라는 메시지를 추가하여 나중에 쉽게 식별할 수 있도록 합니다.
    *   **정기적인 백업**: 중요한 작업 단위를 완료했을 때, 코드 손실을 방지하기 위해 자주 커밋합니다.

    **결론**:
    `git commit`은 **코드를 변경할 때마다 하는 것이 아니라, 의미 있는 변경 단위로 묶어서 정기적으로 하는 것**이 좋습니다. 너무 잦은 커밋은 커밋 로그를 복잡하게 만들고, 너무 드문 커밋은 변경 사항 추적과 협업을 어렵게 할 수 있습니다. **기능 단위로 커밋하고, 명확한 커밋 메시지를 작성**하는 것이 중요합니다. 이를 통해 코드 관리 효율성을 높이고, 문제가 발생했을 때 쉽게 해결할 수 있습니다.

*   **`git log`**: 커밋 히스토리를 확인합니다.
*   **`git checkout`**: 특정 커밋이나 브랜치로 이동합니다.
*   **`git reset`**: 특정 커밋으로 되돌아가며, 변경 사항을 유지하거나 삭제할 수 있습니다. `soft`, `mixed`, `hard` 옵션이 있습니다.
*   **`git revert`**: 특정 커밋을 취소하는 새로운 커밋을 생성합니다. 커밋 히스토리를 유지하면서 변경 사항을 되돌릴 때 사용합니다.
*   **`git stash`**: 현재 변경 사항을 임시로 저장하고 다른 작업을 할 수 있게 해줍니다.
*   **`git merge`**: 한 브랜치의 변경 사항을 다른 브랜치로 병합합니다.
*   **`git pull`**: 원격 저장소의 변경 사항을 로컬 저장소로 가져와 병합합니다.
*  **`git status`**: 현재 브랜치, 커밋 여부, 추적하지 않은 파일 등을 보여줍니다.

**GitHub**:
*   **클라우드 플랫폼**: Git 저장소를 온라인으로 저장하고 팀원과 협업할 수 있게 해줍니다.
*   **원격 저장소**: GitHub, GitLab, Bitbucket과 같은 서버에 저장된 프로젝트 버전입니다.
*   **원격 저장소 연결**: 로컬 저장소를 원격 저장소에 연결하여 변경 사항을 공유하고 협업합니다.
*   **`git remote add origin <url>`**: 로컬 저장소를 원격 저장소에 연결합니다.
*   **`git push`**: 로컬 커밋을 원격 저장소로 업로드합니다.
*   **풀 리퀘스트 (Pull Request)**: 코드 변경 사항을 팀원들과 공유하고 검토를 받기 위한 요청입니다.
*   **병합 충돌 (Merge Conflict)**: 두 개 이상의 브랜치가 같은 파일의 동일한 라인을 수정했을 때 발생하며, 수동으로 해결해야 합니다.

**Git 사용 워크플로우**:
1.  **저장소 복제 (Clone)**: 원격 저장소를 로컬로 가져옵니다.
2.  **새 브랜치 생성**: `main` 또는 다른 브랜치에서 새 브랜치를 만듭니다.
3.  **변경 사항 적용**: 코드 변경 작업을 합니다.
4.  **브랜치 푸시**: 변경된 브랜치를 원격 저장소에 업로드합니다.
5.  **풀 리퀘스트 생성**: 변경 사항을 `main` 브랜치로 병합하기 위해 풀 리퀘스트를 생성합니다.
6.  **변경 사항 병합**: 풀 리퀘스트가 승인되면 `main` 브랜치로 병합합니다.
7.  **로컬 `main` 브랜치 업데이트**: 원격 `main` 브랜치 변경 사항을 로컬 `main` 브랜치로 가져옵니다.

**GUI (Graphical User Interface)를 이용한 Git 작업**:
*   **터미널 명령 없이 Git 작업**: WebStorm과 같은 IDE에서 시각적인 인터페이스를 통해 Git 작업을 수행할 수 있습니다.
*   **버튼, 패널, 메뉴**: 복잡한 명령어를 사용하지 않고 Git 기능을 사용할 수 있습니다.
*   **GUI의 장점**: 효율적인 작업, 쉬운 협업 및 시각적 히스토리 관리가 가능합니다.

**GUI를 사용한 예시**:
* 저장소 초기화, 커밋, 브랜치 생성, 푸시, 풀, 병합, 풀 리퀘스트 생성 및 검토 등 다양한 작업을 GUI를 통해 할 수 있습니다.
